```python
# Q1. Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation of 5 using Python. Interpret the results.

import scipy.stats as stats
import math

mean = 50
std_dev = 5
n = 30  
z_score = stats.norm.ppf(0.975)  
margin_of_error = z_score * (std_dev / math.sqrt(n))
lower_bound = mean - margin_of_error
upper_bound = mean + margin_of_error
print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")

# Interpretation: The confidence interval provides a range in which the true population mean is likely to fall. For the given data, the interval might look something
# like (48.21,51.79), meaning there is a 95% chance that the true population mean is within this range.
```

```python
# Q2. Conduct a chi-square goodness of fit test to determine if the distribution of colors of M&Ms in a bag matches the expected distribution of 20% blue, 
#     20% orange, 20% green, 10% yellow, 10% red, and 20% brown. Use Python to perform the test with a significance level of 0.05.

import scipy.stats as stats

observed_counts = [22, 18, 25, 12, 9, 24]  
total = sum(observed_counts)
expected_counts = [0.2 * total, 0.2 * total, 0.2 * total, 0.1 * total, 0.1 * total, 0.2 * total]
chi2_stat, p_value = stats.chisquare(f_obs=observed_counts, f_exp=expected_counts)
print(f"Chi-Square Statistic: {chi2_stat:.2f}, p-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The observed distribution does not match the expected distribution.")
else:
    print("Fail to reject the null hypothesis: The observed distribution matches the expected distribution.")
```

```python
# Q3. Use Python to calculate the chi-square statistic and p-value for a contingency table with the following data:
#             Group A          Group B
# Outcome1       20              15
# Outcome2       10              25
# Outcome3       15              20

# Interpret the results of the test.

import scipy.stats as stats
data = [[20, 15],  
        [10, 25],  
        [15, 20]]  


chi2_stat, p_value, dof, expected = stats.chi2_contingency(data)
print(f"Chi-Square Statistic: {chi2_stat:.2f}")
print(f"p-value: {p_value:.4f}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies:")
print(expected)

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant association between the groups and outcomes.")
else:
    print("Fail to reject the null hypothesis: No significant association between the groups and outcomes.")


# Interpretation: If the p-value is less than 0.05, reject the null hypothesis, indicating a significant association between the groups (A and B) and outcomes. 
# If the p-value is greater than or equal to 0.05, fail to reject the null hypothesis, indicating no significant association.
```

```python
# Q4. A study of the prevalence of smoking in a population of 500 individuals found that 60 individuals smoked. Use Python to calculate the 95% confidence interval
# for the true proportion of individuals in the population who smoke.

from statsmodels.stats.proportion import proportion_confint

n = 500  
smokers = 60 
confidence_interval = proportion_confint(smokers, n, alpha=0.05, method='normal')
print("95% Confidence Interval:", confidence_interval)
```

```python
# Q5. Calculate the 90% confidence interval for a sample of data with a mean of 75 and a standard deviation of 12 using Python. Interpret the results.

import scipy.stats as stats

mean = 75
std_dev = 12
sample_size = 1  
confidence = 0.90
z_score = stats.norm.ppf((1 + confidence) / 2)
margin_of_error = z_score * std_dev / (sample_size**0.5)
interval = (mean - margin_of_error, mean + margin_of_error)
print("90% Confidence Interval:", interval)

# Interpretation: The true population mean is likely to fall within the calculated interval 90% of the time.
```

```python
# Q6. Use Python to plot the chi-square distribution with 10 degrees of freedom. Label the axes and shade the area corresponding to a chi-square statistic of 15.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

df = 10 
x = np.linspace(0, 30, 500)
y = chi2.pdf(x, df)
plt.plot(x, y, label='Chi-Square Distribution (df=10)')
plt.fill_between(x, y, where=(x <= 15), color='gray', alpha=0.5, label='Area for Chi-Square <= 15')
plt.title('Chi-Square Distribution with Shaded Region')
plt.xlabel('Chi-Square Statistic')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```

```python
# Q8. A researcher hypothesizes that a coin is biased towards tails. They flip the coin 100 times and observe
# 45 tails. Conduct a chi-square goodness of fit test to determine if the observed frequencies match the expected frequencies of a fair coin. Use a significance level of 0.05.

import scipy.stats as stats
observed = [45, 55]  
expected = [50, 50]  
chi2_stat, p_value = stats.chisquare(observed, f_exp=expected)

print("Chi-Square Statistic:", chi2_stat)
print("p-value:", p_value)
if p_value < 0.05:
    print("Reject the null hypothesis: The coin is biased.")
else:
    print("Fail to reject the null hypothesis: No significant evidence of bias.")
```

```python
# Q11. A random sample of 30 people was selected from a population with an unknown mean and standard deviation. The sample mean was found to be 72 and the sample 
# standard deviation was found to be 10. Conduct a hypothesis test to determine if the population mean is significantly different from 70. Use a significance level of 0.05.

from scipy.stats import t

sample_mean = 72
pop_mean = 70
sample_std = 10
sample_size = 30
alpha = 0.05

t_stat = (sample_mean - pop_mean) / (sample_std / (sample_size ** 0.5))
df = sample_size - 1
t_critical = t.ppf(1 - alpha / 2, df)
p_value = 2 * (1 - t.cdf(abs(t_stat), df))

print("T-Statistic:", t_stat)
print("Critical T-Value:", t_critical)
print("P-Value:", p_value)

if abs(t_stat) > t_critical:
    print("Reject the null hypothesis: The population mean is significantly different from 70.")
else:
    print("Fail to reject the null hypothesis: No significant evidence that the population mean differs from 70.")


# Interpretation:
# If the absolute value of the t-statistic exceeds the critical t-value, or if the  p-value is less than 0.05, we reject the null hypothesis and conclude that the population mean is significantly different from 70.
# Otherwise, we fail to reject the null hypothesis, indicating insufficient evidence to conclude a significant difference.
```
